```python
#import libraries
import pandas as pd 
import numpy as np
```

```python
#load dataset into pandas datafram
df = pd.read_csv('moviereviews2.tsv', sep='\t') 
```

```python
df.head() #check the head of the dataframe
```

```python
df.describe() #describe the dataframe 
```

```python
#check for null values in the dataframe
df.isnull().sum() #review column has 20 null values 
```

```python
len(df) #check the length of the dataframe
```

```python
df.dropna(inplace=True) #drop all null values
```

```python
df.isnull().sum() #confirm if null values have been removed 
```

```python
len(df) #reconfirmation by checking the length of the dataframe
```

```python
#check if we have a movie review with white space 
for i in df['review']:
    print(i.isspace())
```

```python
#check if there are empty reviews and store it in an empty array 
blanks = []
for rv in df['review']:
    if rv.isspace():
        blanks.append(rv) 
```

```python

len(blanks) #since blanks has zero element => means there no empty reviews
```

```python
#analyse the label column of the dataframe
df['label'].value_counts()
```

```python
#split data into train and test set
from sklearn.model_selection import train_test_split
X= df['review']

y= df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42) 
```

```python
#pipeline to vectorize data , train and fit model
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
```

```python
#instatiate pipeline 
m_clf = Pipeline([('tfid', TfidfVectorizer()),
                  ('clf', LinearSVC())
                  ])
```

```python
#feed training data through pipeline 
m_clf.fit(X_train, y_train)
```

```python
#form a prediction set 
prediction = m_clf.predict(X_test)  #predict base on X_test so we can compare it with our y_test 
```

```python
#report confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, prediction))
```

```python
#print classification report 
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```

```python
#print accuracy of model 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction))
```
